```python
# adding new features
df_colsDropped['living_lot'] = df_colsDropped.sqft_living + df_colsDropped.sqft_lot
```

```python
from scipy.stats import zscore
df_colsDropped_scaled  = df_colsDropped.apply(zscore)
X = df_colsDropped_scaled.drop(['price'], axis=1)
y = df_colsDropped_scaled.pop('price')
print(X.sample())
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_ = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_, y, shuffle=True, test_size=0.30, random_state=98789)


print(X_train.shape)
print(X_test.shape)
print(y_test.shape)

#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
X_train_std =  X_train #sc.fit_transform(X_train) 
X_test_std =  X_test #sc.fit_transform(X_test) 
```

```python
# feature extraction

from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X, y)

pd.DataFrame(model.feature_importances_, index=X.columns, 
             columns=['Feature Importance']).sort_values('Feature Importance', ascending=False)
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
regression_model.score(X_train, y_train)
```

```python
regression_model.score(X_test, y_test)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

alphas = np.array([100000, 10000, 1000, 100, 1,0.1,0.01,0.001,0.0001, 0.00001, 0.000001, 0])

model = Ridge()
grid = GridSearchCV(estimator=model, param_grid=dict(alpha=alphas))
grid.fit(X, y)

print(grid.best_score_)
print(grid.best_estimator_.alpha)
```

```python
from sklearn import linear_model
regressor = linear_model.Ridge(alpha = grid.best_estimator_.alpha)
regressor.fit(X_train, y_train)
print('The ridge regressor training set accuracy -- {0}'.format(regressor.score(X_train, y_train)))
print('The ridge regressor Test set accuracy -- {0}'.format(regressor.score(X_test, y_test)))
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
alphas = np.array([100000, 10000, 1000, 100,1,0.1,0.01,0.001,0.0001, 0.00001, 0.000001])

model = Lasso()
grid = GridSearchCV(estimator=model, param_grid=dict(alpha=alphas))
grid.fit(X, y)

print(grid.best_score_)
print(grid.best_estimator_.alpha)

```

```python
from sklearn import linear_model
regressor = linear_model.Lasso(alpha = grid.best_estimator_.alpha)
regressor.fit(X_train, y_train)
print('The lasso regressor training set accuracy -- {0}'.format(regressor.score(X_train, y_train)))
print('The lasso regressor test set accuracy -- {0}'.format(regressor.score(X_test, y_test)))
```

After feature selection through Lasso and Ridge, there is slight increase in accuracy.
